```python
import pandas as pd

spotify = pd.read_csv("spotify.csv")

spotify
```

```python
spotify.info()
```

```python
# Convert the 'track_album_release_date' column to datetime, allowing for mixed formats
spotify['track_album_release_date'] = pd.to_datetime(
    spotify['track_album_release_date'], 
    errors='coerce',           # Coerce invalid dates to NaT
    infer_datetime_format=True # Infer format to handle mixed formats
)

# Extract the year from the 'track_album_release_date' column
spotify['track_album_release_year'] = spotify['track_album_release_date'].dt.year

# Filter the DataFrame to include only rows where the release year is 2017
spotify_2017 = spotify[spotify['track_album_release_year'] == 2017]

# Check the resulting DataFrame
spotify_2017
```

```python
# Convert duration from milliseconds to minutes
spotify_2017['duration_ms'] = spotify['duration_ms'] / (1000 * 60)  # Convert milliseconds to minutes

spotify_2017
```

```python
genre2017_list = []

grp_genre2017 = spotify.groupby("playlist_genre")

for grp, grp_info in grp_genre2017:
    genre2017_dict = {}
    genre2017_dict["genre"] = grp
    genre2017_dict["loudness"] = grp_info['loudness'].mean()
    genre2017_dict["mode"] = grp_info['mode'].mode()[0]  # Assuming mode is categorical and taking the first mode
    genre2017_dict["speechiness"] = grp_info['speechiness'].mean()
    genre2017_dict["acousticness"] = grp_info['acousticness'].mean()
    genre2017_dict["instrumentalness"] = grp_info['instrumentalness'].mean()
    genre2017_dict["liveness"] = grp_info['liveness'].mean()
    genre2017_dict["valence"] = grp_info['valence'].mean()
    genre2017_dict["tempo"] = grp_info['tempo'].mean()
    genre2017_dict["duration_m"] = grp_info['duration_ms'].mean()

    genre2017_list.append(genre2017_dict)

# Convert the list of dictionaries to a DataFrame
genre2017_df = pd.DataFrame(genre2017_list)
genre2017_df
```

```python
# Save the DataFrame to a CSV file
genre2017_df.to_csv('genre_2017_averages.csv', index=False)
```
